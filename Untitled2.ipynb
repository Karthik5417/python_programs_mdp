```python
# WAP to find the greatest of two numbers using if and else
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 > num2:
    print(num1, "is greater than", num2)
else:
    print(num2, "is greater than", num1)
```

```python
#wap to find the smallest no. of three no.s using nested if statement

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a < b:
    if a < c:
        print(a, "is the smallest number")
    else:
        print(c, "is the smallest number")
else:
    if b < c:
        print(b, "is the smallest number")
    else:
        print(c, "is the smallest number")
```

```python
#WAP to check whether. a character is a vowel or consonant
ch = input("Enter a character: ")
ch = ch.lower()
if ch.isalpha():
    if ch in ('a', 'e', 'i', 'o', 'u'):
        print(ch, "is a vowel")
    else:
        print(ch, "is a consonant")
else:
    print("Invalid input! Please enter an alphabet.")
```

```python
#wap to check if a year is a leap year or not
year = int(input("Enter a year: "))

if (year % 4== 0):
    print(year, "is a leap year.")
elif (year % 100 == 0):
    print(year, "is not a leap year.")
elif (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
# Program to check if a number is within a specific range

# Take input from the user
num = float(input("Enter a number: "))

# Define the range
lower_limit = 10
upper_limit = 50

# Check if number is in the range
if lower_limit <= num <= upper_limit:
    print(f"{num} is within the range {lower_limit} to {upper_limit}.")
else:
    print(f"{num} is outside the range {lower_limit} to {upper_limit}.")
```
