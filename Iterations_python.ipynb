# iterations or loops in python
1.iterations are used to work with repeatative performances or repeatative tasks
2.iterations are of two types
3. 'for' loop
4. 'while'loop


``` python - for loop syntax
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter variable is type of iterate data from sequence")
    print("in is keyword and operator to check if performance is present or not")
    print("sequence is collection like string,list,tuple,dict,.....etc")
```

``` python - while loop syntax
while (condition):
    print("while is keyword to start the iterations based on conditionds")
    print("if conditions id true then start iterations")
